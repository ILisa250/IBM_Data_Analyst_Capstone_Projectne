# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
WHERE ConvertedComp IS NOT NULL
"""
data = pd.read_sql_query(QUERY, conn)

plt.hist(data['ConvertedComp'], bins=30, edgecolor='black')
plt.title('Distribution of Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT Age
FROM master
WHERE Age IS NOT NULL
"""
data = pd.read_sql_query(QUERY, conn)

plt.boxplot(data['Age'], vert=False)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
WHERE Age IS NOT NULL AND WorkWeekHrs IS NOT NULL
"""
data = pd.read_sql_query(QUERY, conn)

# Scatter plot
plt.scatter(data['Age'], data['WorkWeekHrs'], alpha=0.5)
plt.title('Age vs. Work Hours per Week')
plt.xlabel('Age')
plt.ylabel('Work Hours per Week')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE WorkWeekHrs IS NOT NULL AND CodeRevHrs IS NOT NULL AND Age IS NOT NULL
"""
data = pd.read_sql_query(QUERY, conn)

# Bubble plot
plt.scatter(data['WorkWeekHrs'], data['CodeRevHrs'], s=data['Age'], alpha=0.5)
plt.title('Work Hours vs. Code Review Hours (Bubble size: Age)')
plt.xlabel('Work Hours per Week')
plt.ylabel('Code Review Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""
data = pd.read_sql_query(QUERY, conn)

plt.pie(data['count'], labels=data['DatabaseDesireNextYear'], autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents Wish to Learn')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
AND WorkWeekHrs IS NOT NULL
AND CodeRevHrs IS NOT NULL
"""
data = pd.read_sql_query(QUERY, conn)

median_data = data.groupby('Age').agg(
    MedianWorkWeekHrs=('WorkWeekHrs', 'median'),
    MedianCodeRevHrs=('CodeRevHrs', 'median')
).reset_index()

plt.bar(median_data['Age'], median_data['MedianWorkWeekHrs'], label='Median WorkWeekHrs')
plt.bar(median_data['Age'], median_data['MedianCodeRevHrs'], bottom=median_data['MedianWorkWeekHrs'], label='Median CodeRevHrs')
plt.title('Work Hours and Code Review Hours (Ages 30-35)')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
AND ConvertedComp IS NOT NULL
"""
data = pd.read_sql_query(QUERY, conn)

median_data = data.groupby('Age')['ConvertedComp'].median().reset_index()
median_data.rename(columns={'ConvertedComp': 'MedianComp'}, inplace=True)

plt.plot(median_data['Age'], median_data['MedianComp'])
plt.title('Median Compensation (Ages 45-60)')
plt.xlabel('Age')
plt.ylabel('Median Compensation')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch, COUNT(*) as count
FROM master
GROUP BY MainBranch
ORDER BY count DESC
"""
data = pd.read_sql_query(QUERY, conn)

plt.barh(data['MainBranch'], data['count'])
plt.title('Respondents by Main Branch')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.show()
```

```python
### Rank of Python
QUERY = """
SELECT LanguageDesireNextYear, COUNT(*) as count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY count DESC
"""
data = pd.read_sql_query(QUERY, conn)

data['Rank'] = data['count'].rank(ascending=False, method='min')
python_rank = data[data['LanguageDesireNextYear'] == 'Python']
print(f"Python's rank: {python_rank['Rank'].values[0]}")
```

```python
## Respondents currently working with SQL
QUERY = """
SELECT COUNT(*)
FROM LanguageWorkedWith
WHERE LanguageWorkedWith LIKE '%SQL%'
"""
data = pd.read_sql_query(QUERY, conn)
print(f"Number of respondents currently working with SQL: {data.iloc[0, 0]}")
```

```python
## Respondents working on MYSQL Only
QUERY = """
SELECT COUNT(*)
FROM LanguageWorkedWith
WHERE LanguageWorkedWith = 'SQL'
"""
data = pd.read_sql_query(QUERY, conn)
print(f"Number of respondents working only with MySQL: {data.iloc[0, 0]}")
```

```python
### Listing all available languages 
QUERY = """
SELECT DISTINCT LanguageWorkedWith AS Language
FROM LanguageWorkedWith
ORDER BY Language
"""
data = pd.read_sql_query(QUERY, conn)

print("Available Languages (Current Work):")
print(data['Language'].tolist())

QUERY = """
SELECT DISTINCT LanguageDesireNextYear AS Language
FROM LanguageDesireNextYear
ORDER BY Language
"""
data = pd.read_sql_query(QUERY, conn)

print("\nAvailable Languages (Desired Next Year):")
print(data['Language'].tolist())
```
